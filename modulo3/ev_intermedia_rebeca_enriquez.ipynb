- Haciendo el an√°lisis exploratorio te deber√≠as haber dado cuenta de que algunas de las variables (`danceability`, `energy`, `key`, `loudness`, `speechiness`, `acousticness`, `instrumentalness`) no son del tipo que deber√≠an. Esto es debido a que los decimales est√°n establecidos como comas y no con puntos. Crea una funci√≥n que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto. 

- Despu√©s de haber hecho los cambios, chequea los tipos de datos. ¬øSon ya del tipo correcto? En caso de que no, crea otra funci√≥n o modifica la anterior para que sean de tipo *float*.

```python
def cambio_signos_puntuacion(elemento):

    try:     
        return float(elemento.replace(",","."))
    
    except:
        return np.nan
```

```python
columnas_cambios = ["danceability", "energy", "key", "loudness", "speechiness", "acousticness", "instrumentalness"]
```

```python
for columna in columnas_cambios:
    df[columna] = df[columna].apply(cambio_signos_puntuacion)
```

```python
for columna in columnas_cambios:
    print(f"la columnas {columna.upper()} es de tipo:{df[columna].dtype}")

# Podemos hacer df.dtypes y ahorrarnos el bucle
```

- Hay algunas canciones cuyo t√≠tulo est√° en may√∫sulas. Crea una funci√≥n para que todos los t√≠tulos est√©n en min√∫scula. Haz lo mismo para las columnas de "title" y "album".


```python
columnas_cambios_miniscula =["track", "title", "album"]
```

```python
for columna in columnas_cambios_miniscula:
    df[columna] = df[columna].str.lower()
```

Lo podemos hacer sobre el DF como se puede ver arriba o haciendo una funcion y aplicando a cada celda con apply

```python
def cambio_minusculas(celda):
    try:
        return celda.lower()
    except:
        return None
```

```python
for columna in columnas_cambios_miniscula:
    df[columna] = df[columna].apply(cambio_minusculas)
```

```python
# Para hacerlo con una lambda
for col in columnas_cambios_miniscula:
    df[col] = df[col].apply(lambda dato: dato.lower() if type(dato)== str else dato)
```

- Algunos de los artistas tienen s√≠mbolos raros, en concreto el s√≠mbolo `$`. Crea una funci√≥n que los reemplaze pos "s". 

```python
# hacer la comprobacion
# Sin regex
df[df["artist"].str.contains("$", regex = False)]["artist"].unique()

# Con regex
df[df["artist"].str.contains("\$", regex = True)]["artist"].unique()
```

```python
columnas_cambios_simbolo = ["artist"]
```

```python
def cambio_simbolo(celda):
    return celda.replace("$","s")
```

```python
for columna in columnas_cambios_simbolo:
    df[columna] = df[columna].apply(cambio_simbolo)
```

```python
# Lo podemos hacer con una lambda
df["artist"] = df["artist"].apply(lambda dato: dato.replace("$","S"))

# Comprobamos que nos ha cambiado los datos, nuestro objetivo es que la lista que devuelva este vacia
df[df["artist"].str.contains("$", regex = False)]["artist"].unique()
```
